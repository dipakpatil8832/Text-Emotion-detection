```python
# 1. Add Student
# 2. View Students
# 3. Search Student
# 4. Update Marks
# 5. Delete Student
# 6. Exit
```

```python
name_data=[]
marks_data=[]
roll_num_data=[]
```

```python

data = {"roll_num":roll_num_data,
        "name":name_data,
        "marks":marks_data}
```

```python

def add_student():
  name = input("enter student name : ")
  roll = int(input("enter roll number : "))
  marks = int(input("enter marks : "))
  if roll > 0 and roll not in roll_num_data and name.isalpha() and marks > 0:
    name_data.append(name)
    roll_num_data.append(roll)
    marks_data.append(marks)
  else :
    if  not name.isalpha():
      print("name should be text")
    elif roll < 0 or roll in roll_num_data :
      print("!!! roll numer allready taken")
    elif marks < 0 :
      print("marks should be greater than 0")
    else :
      print("please enter valid details")
  return "--"*25

def View_Students():
  import pandas as pd
  df = pd.DataFrame(data)
  print(df)

def Search_Student():
  id= int(input("enter roll number"))
  if id in roll_num_data:
    for i in data :
      if i == "roll_num":
        for j in data[i] :
          if j == id:
            index = roll_num_data.index(id)

            print("name     : ",name_data[index])
            print("roll num : ",roll_num_data[index])
            print("marks    : ",marks_data[index])

  else :
    print("student not found !!!")
  return "--"*25

def Update_Marks():
  id = int(input("enter roll number for update marks: "))
  if id in roll_num_data:
    up_marks=int(input("enter new marks: "))
    for i in data :
      if i == "roll_num":
        for j in data[i] :
          if j == id:
            index = roll_num_data.index(id)
            marks_data[index]=up_marks
  else:
      print("student not found !!!")
  return "--"*25


def Delete_Student():
  id= int(input("enter roll number"))
  if id in roll_num_data:
    index = roll_num_data.index(id)
    name_data.remove(name_data[index])
    marks_data.remove(marks_data[index])
    roll_num_data.remove(roll_num_data[index])
  else :
    print("student not found !!!")
  return "--"*25


print("""1. Add Student
2. View Students
3. Search Student
4. Update Marks
5. Delete Student
6. Exit""")
choice = int(input("enter choice: "))
while choice != 6:

  if choice == 1:
    add_student()
  elif choice == 2:
    View_Students()
  elif choice == 3:
    Search_Student()
  elif choice == 4:
    Update_Marks()
  elif choice == 5:
    Delete_Student()
  else:
    print("invalid choice")

  print("--"*20)
  print("""1. Add Student
2. View Students
3. Search Student
4. Update Marks
5. Delete Student
6. Exit""")
  choice = int(input("enter choice: "))




```
